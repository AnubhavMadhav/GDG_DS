## Importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

## Fetching HTML content for a given URL

```python
r = requests.get("https://www.olx.in/vadodara_g4058732/q-car")
print(r)
```

## Parsing the HTML content

```python
soup = BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python
l=[]
box_data = soup.find_all(attrs={"data-aut-id": "itemBox"})
for item in box_data:
    dict={}
    
    vehicles = item.find(attrs={"data-aut-id": "itemTitle"})
    dict["Vehicle"]= vehicles.text.split(',',1)[0]
    loc=item.find(attrs={"data-aut-id": "item-location"})
    dict["Location"]=loc.text.split(',',1)[0]
    pricekms = item.find(attrs={"data-aut-id": "itemPrice"})
    dict["Price"]= pricekms.text
    
    details = item.find(attrs={"data-aut-id": "itemDetails"})
    try:
        dict["Kms"]= details.text.split('-',1)[1]
        dict["Model"]=details.text.split('-',1)[0]
        
    except:
        dict["Kms"]="-"
        dict["Model"]="-"
    
    l.append(dict)
```

## Creating a DataFrame

```python
import pandas
df= pandas.DataFrame(l)
```

## Writing DataFrame into a CSV file

```python
df.to_csv("beautiful_soup_demo.csv")
```
